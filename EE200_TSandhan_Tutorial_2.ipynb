***Signals*** - A signal is a function of one or more variables that indicate some (usually physical) phenomenon. Signal serves as carriers of information between communication devices. They can convey different types of information depending on the application required. These signals can be of different forms.


***Different Type of Signals***



1.   Unit Step Signal
2.   Ramp Signals
3.   Triangular Signals
4.   Rectangular Signal
5.   Impulse Signal



```python
from IPython.display import Markdown
# Cell 2: Markdown
Markdown(r'''
## ðŸ”¹ Unit Step Signal

**Theory**
\[
u(t) = \begin{cases}
0, & t < 0 \\
1, & t \geq 0
\end{cases}
\]
A unit step function switches a signal ON at \( t = 0 \).
''')
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
u = np.heaviside(t, 1)  # 1 at t=0

plt.plot(t, u, label='Unit Step')
plt.title('Unit Step Signal')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

```python
from IPython.display import Markdown
# Cell 4: Markdown
Markdown(r'''
## ðŸ”¹ Ramp Signal

**Theory**
\[
r(t) = t . u(t)
\]
A linearly increasing function starting from zero.
''')
```

```python
# Cell 5: Ramp Signal
r = t * np.heaviside(t, 1)

plt.plot(t, r)
plt.title('Ramp Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```


***Triangular Signal***
A triangular signal has a peak and falls symmetrically on both sides like a triangle.

```python
tri = 1 - np.abs(t)/5
tri[t < -5] = 0
tri[t > 5] = 0

plt.plot(t, tri, label='Triangular Signal')
plt.title('Triangular Signal')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```


***Square Signal***
A square wave alternates between +1 and -1 with a 50% duty cycle.

```python
from scipy import signal

square = signal.square(2 * np.pi * 1 * t)

plt.plot(t, square, label='Square Wave')
plt.title('Square Wave Signal')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

***Sinusoidal Signal***
A sine wave is a smooth periodic oscillation defined as
Asin(2Ï€ft).

```python
sine = 3*np.sin(2 * np.pi * 1 * t)

plt.plot(t, sine, label='Sine Wave')
plt.title('Sine Wave')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

***Cosine Signal***
A cosine wave is similar to sine but starts at maximum amplitude:
Acos(2Ï€ft).

```python
cosine = np.cos(2 * np.pi * 1 * t)

plt.plot(t, cosine, label='Cosine Wave', color='orange')
plt.title('Cosine Wave')
plt.grid()
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.show()
```

# **Addition & Scaling Properties of Signals**
Signals can be added or scaled just like vectors. Linearity is a fundamental property in signal processing.

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(0, 2, 1000)  # 2 seconds duration

# Lower frequencies for better visualization
f1 = 1  # Hz
f2 = 2  # Hz

# Signals
s1 = np.sin(2 * np.pi * f1 * t)
s2 = 0.5 * np.sin(2 * np.pi * f2 * t)
combined = s1 + s2

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(t, s1, label='s1: 1 Hz', linestyle='-', color='blue')
plt.plot(t, s2, label='s2: 2 Hz (scaled by 0.5)', linestyle='--', color='green')
plt.plot(t, combined, label='Combined Signal', linestyle='-', color='red')

plt.title('Signal Addition and Scaling (1 Hz + 0.5Ã—2 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

# ***Time Shifting and Reversal***

Shifting moves the signal in time:

x(t-t

Reversal reflects the signal:

x(-t).

```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
t = np.linspace(-1, 1, 1000)

# Use a lower frequency for better visualization
frequency = 2  # Hz

# Define signals
original = np.sin(2 * np.pi * frequency * t)
shifted = np.sin(2 * np.pi * frequency * (t - 0.2))
reversed_signal = np.sin(2 * np.pi * frequency * (-t))

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(t, original, label='Original', color='blue')
plt.plot(t, shifted, label='Shifted (0.2s)', linestyle='--', color='green')
plt.plot(t, reversed_signal, label='Reversed', linestyle=':', color='red')

plt.title('Time Shifting and Reversal ({} Hz Sine Wave)'.format(frequency))
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

## Question 1

A discrete-time signal \( x[n] \) is defined as:

\[
x[n] = 
\begin{cases}
n + 2, & -2 \leq n < 0 \\
2 - n, & 0 \leq n \leq 2 \\
0, & \text{otherwise}
\end{cases}
\]

###  Tasks:

1. Plot the discrete signal \( x[n] \).
2. Compute the **even** and **odd** components:
   - Even part: 
     \[
     x_{\text{even}}[n] = \frac{x[n] + x[-n]}{2}
     \]
   - Odd part: 
     \[
     x_{\text{odd}}[n] = \frac{x[n] - x[-n]}{2}
     \]
3. Plot both even and odd components of the signal.


```python
import numpy as np
import matplotlib.pyplot as plt

n = np.arange(-5, 6)

# Define x[n] manually
x = np.zeros_like(n, dtype=float)

for i, val in enumerate(n):
    if -2 <= val < 0:
        x[i] = val + 2
    elif 0 <= val <= 2:
        x[i] = 2 - val

x_neg_n = np.zeros_like(n, dtype=float)
for i, val in enumerate(n):
    if -val in n:
        x_neg_n[i] = x[np.where(n == -val)[0][0]]

x_even = 0.5 * (x + x_neg_n)
x_odd = 0.5 * (x - x_neg_n)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(n, x, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title('Discrete-Time Signal x[n]')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.stem(n, x_even, basefmt=" ", linefmt='g-', markerfmt='go')
plt.title('Even Component x_even[n]')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.stem(n, x_odd, basefmt=" ", linefmt='r-', markerfmt='ro')
plt.title('Odd Component x_odd[n]')
plt.grid(True)

plt.tight_layout()
plt.show()
```

**Question 3**: A complex signal is defined as:

x(t)=e^j2Ï€ft

where
f=2 Hz.

Plot the real part of x(t).

Plot the imaginary part of x(t).

Plot the magnitude and phase of x(t).



```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
f = 2

x_t = np.exp(1j * 2 * np.pi * f * t)

real_part = np.real(x_t)
imag_part = np.imag(x_t)
magnitude = np.abs(x_t)
phase = np.angle(x_t)

plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(t, real_part)
plt.title("Real Part of x(t) = cos(2Ï€ft)")
plt.xlabel("Time (t)"); plt.ylabel("Re{x(t)}")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(t, imag_part)
plt.title("Imaginary Part of x(t) = sin(2Ï€ft)")
plt.xlabel("Time (t)"); plt.ylabel("Im{x(t)}")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(t, magnitude)
plt.title("Magnitude of x(t)")
plt.xlabel("Time (t)"); plt.ylabel("|x(t)|")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(t, phase)
plt.title("Phase of x(t)")
plt.xlabel("Time (t)"); plt.ylabel("âˆ x(t) [radians]")
plt.grid(True)

plt.tight_layout()
plt.show()
```

##  Question 4:

Given a real-valued signal:

$
x(t) = t \cdot e^{-|t|}
$

1. Find the even part $ x_{\text{even}}(t) $ of the signal using the formula:

$
x_{\text{even}}(t) = \frac{x(t) + x(-t)}{2}
$

2. Find the odd part $ x_{\text{odd}}(t) $ using:

$
x_{\text{odd}}(t) = \frac{x(t) - x(-t)}{2}
$

3. Verify that:

$
x(t) = x_{\text{even}}(t) + x_{\text{odd}}(t)
$

4. Plot all three signals and comment on their symmetry properties.

```python
import matplotlib.pyplot as plt

t = np.linspace(-5, 5, 1000)
x = t * np.exp(-np.abs(t))
x_neg = (-t) * np.exp(-np.abs(-t))
x_even = (x + x_neg) / 2
x_odd = (x - x_neg) / 2
x_check = x_even + x_odd

plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
plt.plot(t, x, label='x(t)')
plt.title("Original Signal x(t)")
plt.xlabel("t"); plt.ylabel("x(t)")
plt.grid(True); plt.legend()

plt.subplot(2, 2, 2)
plt.plot(t, x_even, label='x_even(t)', color='green')
plt.title("Even Part of x(t)")
plt.xlabel("t"); plt.ylabel("x_even(t)")
plt.grid(True); plt.legend()

plt.subplot(2, 2, 3)
plt.plot(t, x_odd, label='x_odd(t)', color='orange')
plt.title("Odd Part of x(t)")
plt.xlabel("t"); plt.ylabel("x_odd(t)")
plt.grid(True); plt.legend()

plt.subplot(2, 2, 4)
plt.plot(t, x, label='x(t)', linestyle='dotted')
plt.plot(t, x_check, label='x_even(t) + x_odd(t)', linestyle='--')
plt.title("Verification: x(t) = x_even(t) + x_odd(t)")
plt.xlabel("t"); plt.ylabel("Values")
plt.grid(True); plt.legend()

plt.tight_layout()
plt.show()
```
